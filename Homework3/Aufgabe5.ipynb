```python
# Implementation der Ausgleichsgerade durch least-squares:

import numpy as np
import matplotlib.pyplot as plt

values = np.loadtxt('vl5_data.csv')

t_u = np.array(values[:,1])
t_c = np.array(values[:,0])

mean_u = np.mean(t_u)
mean_c = np.mean(t_c)

alpha_1 = (np.sum(((t_u-mean_u)*(t_c-mean_c))))/(np.sum((t_u-mean_u)**2))

alpha_0 = mean_c - alpha_1*mean_u

x_vektor = range(15,90)

f_gerade = alpha_0 + x_vektor*alpha_1


plt.plot(t_u,t_c,'x')
plt.plot(x_vektor,f_gerade)
plt.title('Approximation durch least squares')
plt.legend(['originale Werte','Approximation'])

```
